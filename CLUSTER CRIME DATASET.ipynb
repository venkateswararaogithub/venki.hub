```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import fcluster
```

```python
data= pd.read_csv("crime_data.csv")
```

```python
data
```

```python
data.info()
```

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)
data
```

```python
data_norm=StandardScaler().fit_transform(data)
data_norm
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(data_norm)
```

```python
dbscan.labels_
```

```python
data['clusters']=dbscan.labels_
data
```

```python
data.groupby('clusters').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data['clusters'],data['UrbanPop'], c=dbscan.labels_) 
```

```python
def norm_fun(i):
    X=(i-i.min())/(i.max()-i.min())
    return (X)
```

```python
df_n=norm_fun(data.iloc[:,1:])
df_n.head()
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_n, method='single'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_n, method='complete'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_n, method='average'))
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_n, method='centroid'))
```

```python
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average')
hc
```

```python
y_hc=hc.fit_predict(df_n)
clusters=pd.DataFrame(y_hc,columns=['cluster_id'])
clusters
```

```python
data['h_cluster_id']=hc.labels_
```
